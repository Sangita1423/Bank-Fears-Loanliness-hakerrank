## Handle missing values

```python
sns.heatmap(data1.isnull(), yticklabels= False, cbar=False, cmap='YlGnBu')
```

```python
data1['batch_enrolled'].fillna(data1['batch_enrolled'].mode(), inplace = True)
data1['emp_title'].fillna(data1['emp_title'].mode(), inplace = True)
data1['emp_length'].fillna(data1['emp_length'].mode(), inplace = True)
data1['annual_inc'].fillna(data1['annual_inc'].median(), inplace = True)
data1['batch_enrolled'].fillna(data1['batch_enrolled'].mode(), inplace = True)
data1['sub_grade'].fillna(data1['sub_grade'].mode(), inplace = True)
data1['revol_util'].fillna(data1['revol_util'].median(), inplace = True)
data1['total_acc'].fillna(data1['total_acc'].median(), inplace = True)
data1['collections_12_mths_ex_med'].fillna(data1['collections_12_mths_ex_med'].median(), inplace = True)
```

```python
data1['title']=data1['title'].fillna(data1['title'].mode()[0])
data1['delinq_2yrs']=data1['delinq_2yrs'].fillna(data1['delinq_2yrs'].mode()[0]) 
data1['inq_last_6mths']=data1['inq_last_6mths'].fillna(data1['inq_last_6mths'].mode()[0])
data1['tot_coll_amt']=data1['tot_coll_amt'].fillna(data1['tot_coll_amt'].mode()[0])
data1['acc_now_delinq']=data1['acc_now_delinq'].fillna(data1['acc_now_delinq'].mode()[0])
data1['tot_cur_bal']=data1['tot_cur_bal'].fillna(data1['tot_cur_bal'].mode()[0]) 
data1['total_rev_hi_lim']=data1['total_rev_hi_lim'].fillna(data1['total_rev_hi_lim'].mode()[0])
 
```

```python
data1.drop(['desc'], axis=1, inplace=True)
data1.drop(['mths_since_last_delinq'], axis=1, inplace=True)
data1.drop(['mths_since_last_record'], axis=1, inplace=True)
data1.drop(['mths_since_last_major_derog'], axis=1, inplace=True)
data1.drop(['verification_status_joint'], axis=1, inplace=True)
```

```python
data1.dropna(inplace=True)
```
